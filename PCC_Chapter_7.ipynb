```python
#7-5
while True:
  #lo convertimos en 'lower' para que funcione
  #incluso con mayúsculas
  age = input("Cual es su edad: ").lower().strip()
  if age == 'q': break
  #Despues de verificar if 'q', convertimos 'age' en int
  #para que no haya errores
  age = int(age)
  if age < 3: print("The ticket's free!")
  elif age < 13: print("The price of the ticket is $10")
  else: print("The price of the ticket is $15")
```

```python
#7-8
sandwich_orders = ["ham", "tuna", "turkey"]
finished_sandwiches = []

while True:
  if sandwich_orders:
    current = sandwich_orders.pop()
    print(f"I made your {current.title()} sandwich!")
    finished_sandwiches.append(current)
  else: break

print("\nThese are the made sandwiches:")
for sandwich in finished_sandwiches:
  print(f"\t{sandwich.title()} sandwich")
```

```python
#7-9
sandwich_orders = ["pastrami", "ham", "tuna", "pastrami", "turkey", "pastrami"]
finished_sandwiches = []

while 'pastrami' in sandwich_orders:
  sandwich_orders.remove('pastrami')
  continue

while True:
  if sandwich_orders:
    current = sandwich_orders.pop()
    print(f"I made your {current.title()} sandwich!")
    finished_sandwiches.append(current)
  else: break

print("\nThese are the made sandwiches:")
for sandwich in finished_sandwiches:
  print(f"\t{sandwich.title()} sandwich")
```

```python
#7-10
responses = {}
while True:
  name = input("Nombre: ")
  choice = input(f"{name.title()}, if you could visit one place in the world,"
  "where would you go? ")
  #Esta es la forma de guardar 2 inputs como key y value en nuestro diccionario
  #vacio
  responses[name] = choice
  print("Gracias por responder!")
  next = input("Wanna pass the poll to someone else?(y/n) ").lower().strip()
  if next == 'n':
    break

for k, v in responses.items():
  print(f"\n{k.title()} would love to go to {v.title()}")
```
