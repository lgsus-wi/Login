```python
#8-7
def make_album(artist, album, q_songs=None):
  if q_songs:
    dic = {'name':artist,
          'album':album,
           '# de temas':q_songs,
          }
  else:
    dic = {'name':artist,
          'album':album,
          }
  return dic
art1 = make_album('ed sheeran','divide', 14)
art2 = make_album('shakira','waka', 9)
art3 = make_album('chayanne','torero')
print(art1)
print(art2)
print(art3)
```

```python
#8-8
#REPASAR IMPORTANTE:  ASI SE USAN FUNCIONES CON WHILE LOOP
def make_album(artist, album, q_songs=None):
  if q_songs:
    info = {'artist':artist, 'album':album, '# temas':q_songs}
  else: info = {'artist':artist, 'album':album}
  return info

while True:
  art_name = input("Nombre del artista: ")
  if art_name == 'q':
     break
  alb_name = input("Nombre del album: ")
  if alb_name == 'q':
     break
  songs_number = int(input("Cuantas canciones hay en el album: "))
  if songs_number == 'q':
     break

  #Se le otorga variable a nuestro return de la funcion
  dic = make_album(art_name, alb_name, songs_number)
  print(dic)
```

```python
#8-10
words = ['hola', 'como', 'estan', '?']
sent_messages = []

def send_messages(send, sent):
  while send:
    current = send.pop()
    print(current)
    sent.append(current)

send_messages(words, sent_messages)
print(words)
print(sent_messages)
```

```python
#8-11
words = ['hola', 'como', 'estan', '?']
sent_messages = []

def send_messages(send, sent):
  while send:
    current = send.pop()
    print(current)
    sent.append(current)

send_messages(words[:], sent_messages)
print(words)
print(sent_messages)
```

```python
#8-12
def sandwich_toppings(*toppings):
  print("Estos son los agregados para tu sandwich: ")
  for topping in toppings:
    print(f"\t{topping.title()}")

sandwich_toppings('cheese', 'ham', 'lettuce')
```

```python
#8-13
def build_profile(first, last, **user_info):
  user_info['first'] = first.title()
  user_info['last'] = last.title()
  return user_info

user_info = build_profile('jesus','range',age=29,nacionality='colombian',ethnicity='hispanic')
print(user_info)
```

```python
#8-14
def build_car(manufacturer, model, **details):
  details['manufacturer'] = manufacturer
  details['model'] = model
  return details

car = build_car('la 8', 'ford', color='blue', year=2014)

print(car)
```
